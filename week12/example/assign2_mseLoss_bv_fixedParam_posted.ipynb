### Set Up Data

```python
df = cleaned_df.copy()
```

```python
# Create list of input attributes and output variable
df = cleaned_df.copy()
output_var = 'Calories'
input_vars = df.columns.tolist()
input_vars.remove('Food Name')
input_vars.remove('Category Name')
input_vars.remove('Calories')
# Create attribute input and output datasets
X = df[input_vars]
y = df[output_var]
```

```python
import statsmodels.api as sm
X = sm.add_constant(X, prepend=False)
```

```python
# Change number of parameters here
num_params = 29
```

### For a specific groundtruth y, Create (1) Distribution of f_hat, and find (2) a Value of E(f_hat)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
test_percent = 0.2
main_X_train, X_test, main_y_train, y_test = train_test_split(X, y, test_size=test_percent) 

from sklearn.preprocessing import StandardScaler
scaled_main_X_train = StandardScaler().fit_transform(main_X_train)
scaled_X_test = StandardScaler().fit_transform(X_test)

num_iterations = 1000     # can choose other  values

selected_percent = [0.1, 0.2, 0.3]   # can choose other  values
selected_train_size = [int(len(main_X_train)*percent) for percent in selected_percent]
print(selected_train_size)
test_size = len(X_test)

test_f_hat = np.zeros((len(selected_train_size), num_iterations))  

# select the first value in y_test to be ground truth (can select any value in y_test)
ground_truth_y = y_test.tolist()[0]  
print(ground_truth_y)

for t in range(len(selected_train_size)):
    train_size = selected_train_size[t]
    for i in range(num_iterations):                 
        rand_index = np.random.permutation(len(main_X_train))[:train_size]
        idx = main_y_train.index[rand_index]
        scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train.loc[idx]

        param_X_train = scaled_X_train[:,0:num_params]
        param_X_test = scaled_X_test[:,0:num_params]
        
        lm = LinearRegression()
        lm.fit(param_X_train, y_train)
        test_param_y_predicted = lm.predict(param_X_test)
        test_f_hat[t,i] = test_param_y_predicted[0]  # prediction of ground truth y

# test_f_hat dimension = len(selected_train_size) x num_iterations
# np.mean(W,d) = average W in d dimension, where d = 0 = first dimension 
test_E_f_hat = np.mean(test_f_hat, 1)  # average over 2nd dimension or num_iterations
                                       # test_E_f_hat dimension = len(selected_train_size)
print(test_E_f_hat)
```

```python
# Plot distribution of f_hat for 3 parameters, along with E_f_hat and ground_truth_y
max_heights = [0.005, 0.15, 0.15]   # can choose other  values
colors = np.array(['tab:green', 'tab:purple', 'tab:orange'])

fig, axs = plt.subplots(len(selected_train_size), figsize=(12, 6))
for k in range(len(selected_train_size)):    
    # plot distribution of f_hat
    axs[k].hist(test_f_hat[k], density=True, color=colors[k], alpha=0.6, label=r'$\hat{f}$')   
    
    #plot E_f_hat
    axs[k].plot([test_E_f_hat[k], test_E_f_hat[k]], [0, max_heights[k]], c='k', linewidth=2.0, label=r'$\mathbb{E}[\hat{f}]$')

    # plot ground_truth_y
    axs[k].plot([ground_truth_y, ground_truth_y], [0, max_heights[k]], 'r', linewidth=2.0, label=r'$ground\_truth\_y$')

    # set title and legend
    axs[k].title.set_text('#Parameters = {0}, Train Percent = {1:.0f}%'.format(num_params, selected_percent[k]*100))
    axs[k].legend(fontsize=12)
    
plt.show()
```

### Bias-Variance Tradeoff

```python
from sklearn.model_selection import train_test_split
test_percent = 0.2
main_X_train, X_test, main_y_train, y_test = train_test_split(X, y, test_size=test_percent) 

from sklearn.preprocessing import StandardScaler
scaled_main_X_train = StandardScaler().fit_transform(main_X_train)
scaled_X_test = StandardScaler().fit_transform(X_test)

num_iterations = 1000     # can choose other  values

interval = 0.05       # can choose other  values
start_percent = 0.2   # can choose other  values
end_percent = 1       # can choose other  values 
train_percent_list = np.arange(start_percent, end_percent, interval)
train_size_list = [int(len(main_X_train)*percent) for percent in train_percent_list]
print(train_size_list)

test_size = len(X_test)
print(test_size)

train_squared_error = np.zeros((len(train_size_list), num_iterations))  # shape: Percents x Iterations
test_f_hat = np.zeros((len(train_size_list), num_iterations, test_size))  # shape: Percents x Iterations x Test_size

for t in range(len(train_size_list)):
    train_size = train_size_list[t]
    
    for i in range(num_iterations):
        # Shuffle and select training set with specific train_size
        rand_index = np.random.permutation(len(main_X_train))[:train_size]  
        idx = main_y_train.index[rand_index]
        scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train.loc[idx]
    
        # Get the leftmost num_params columns to be inputs of linear model
        param_X_train = scaled_X_train[:,0:num_params]
        param_X_test = scaled_X_test[:,0:num_params]

        # Derive the model with the specified inputs
        lm = LinearRegression()
        lm.fit(param_X_train, y_train)

        # Make predictions for training and test sets
        train_param_y_predicted = lm.predict(param_X_train)
        test_param_y_predicted = lm.predict(param_X_test)

        # Compute MSE for all samples in training set
        train_squared_error[t, i] = np.mean((y_train.tolist() - train_param_y_predicted) ** 2) # shape: Percents x Iterations

        # Collect predictions of test sets to compute MSE for test sets later
        test_f_hat[t, i, :] = test_param_y_predicted  # shape: Percents x Iterations x Test_size

print('Done')
```

```python
# train_squared_error dimension = # shape: Percents x Iterations
# Get average MSE for all iterations
avg_train_squared_error = np.mean(train_squared_error, 1)  # shape: Percents

# test_f_hat dimension = Percents x Iterations x Test_size
# Get average f_hat = E[f_hat] by avearaing over iterations
test_E_f_hat = np.mean(test_f_hat, 1)   # shape: Percents x Test_size

# test_f_hat dimension = Percents x Iterations x Test_size
# Compute test_squared_error = ||y - f_hat ||^2  for each iteration, for each sample in the test set
test_all_squared_error = (y_test.tolist() - test_f_hat) ** 2  # shape: Percents x Iterations x Test_size
# Averaging test_squared_error for all iterations, for each sample in the test set  
test_sample_squared_error = np.mean(test_all_squared_error, 1)   # shape: Percents x Test_size
# Compute average test_squared_error for all samples in the test set
avg_test_squared_error = np.mean(test_sample_squared_error, 1)    # shape: Percents

# test_E_f_hat dimension = Percents x Test_size
# Compute bias_squared = || y - E[f_hat] ||^2 for each sample in the test set
test_sample_bias_squared = (y_test.tolist() - test_E_f_hat) ** 2   # shape: Percents x Test_size
# Compute average bias_squared for all samples in the test set
avg_test_bias_squared = np.mean(test_sample_bias_squared, 1)  # shape: Percents

# test_f_hat dimension = Percents x Iterations x Test_size
# Compute Var[f_hat] for each sample in the test set
test_sample_var_f_hat = np.var(test_f_hat, 1)   # shape: Percents x Test_size
# Compute average bias_squared for all samples in the test set
avg_test_var_f_hat = np.mean(test_sample_var_f_hat, 1)  # shape: Percents

print(avg_train_squared_error)
print(avg_test_squared_error)
print(avg_test_bias_squared)
print(avg_test_var_f_hat)
```

```python
# Plot bias-variance 
x_list = [percent*100 for percent in train_percent_list]
xtick_labels = ['{0:.0f}'.format(percent*100) for percent in train_percent_list]

fig, ax = plt.subplots(3,1)
ax[0].plot(x_list, avg_test_bias_squared, 'y-.', label=r'bias squared: $(\mathbb{E}[\hat{f}] - y)^2$')  
ax[0].set_xticks(x_list)
ax[0].set_xticklabels(xtick_labels, rotation=60)
ax[0].set_xlabel('Percent of Training Set', size=10)
ax[0].set_ylabel('Squared Error', size=10)
ax[0].legend(fontsize=10)
ax[0].set_title('Number of Parameters = '+str(num_params), fontsize=10)

ax[1].plot(x_list, avg_test_var_f_hat, 'b:', label=r'$var(\hat{f})$')  
ax[1].set_xticks(x_list)
ax[1].set_xticklabels(xtick_labels, rotation=60)
ax[1].set_xlabel('Percent of Training Set', size=10)
ax[1].set_ylabel('Squared Error', size=10)
ax[1].legend(fontsize=10)

ax[2].plot(x_list, avg_test_bias_squared, 'y-.', label=r'bias squared: $(\mathbb{E}[\hat{f}] - y)^2$')  
ax[2].plot(x_list, avg_test_var_f_hat, 'b:', label=r'$var(\hat{f})$')  
ax[2].set_xticks(x_list)
ax[2].set_xticklabels(xtick_labels, rotation=60)
ax[2].set_xlabel('Percent of Training Set', size=10)
ax[2].set_ylabel('Squared Error', size=10)
ax[2].legend(fontsize=10)

plt.show()
```
