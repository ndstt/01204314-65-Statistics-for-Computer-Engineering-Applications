### Set Up Data

```python
# Create list of input attributes and output variable
df = cleaned_df.copy()
output_var = 'Calories'
input_vars = df.columns.tolist()
input_vars.remove('Food Name')
input_vars.remove('Category Name')
input_vars.remove('Calories')
# Create attribute input and output datasets
X = df[input_vars]
y = df[output_var]
```

```python
import statsmodels.api as sm
X = sm.add_constant(X, prepend=False)
```

### For a specific groundtruth y, Create (1) Distribution of f_hat, and find (2) a Value of E(f_hat)

```python
len(input_vars)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
test_percent = 0.2
main_X_train, X_test, main_y_train, y_test = train_test_split(X, y, test_size=test_percent) 

from sklearn.preprocessing import StandardScaler
scaled_main_X_train = StandardScaler().fit_transform(main_X_train)
scaled_X_test = StandardScaler().fit_transform(X_test)

num_iterations = 1000   # can choose other values
selected_params = [5, 15, 25]   # can choose other paramater values

train_size = int(len(main_X_train)*train_percent)
test_size = len(X_test)

test_f_hat = np.zeros((len(selected_params), num_iterations))  

# select the first value in y_test to be ground truth (can select any value in y_test)
ground_truth_y = y_test.tolist()[0]  

for i in range(num_iterations):         
    rand_index = np.random.permutation(len(main_X_train))[:train_size]
    idx = main_y_train.index[rand_index]
    scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train.loc[idx]

    for p in range(len(selected_params)):
        num_params = selected_params[p]
        param_X_train = scaled_X_train[:,0:num_params]
        param_X_test = scaled_X_test[:,0:num_params]
        
        lm = LinearRegression()
        lm.fit(param_X_train, y_train)
        test_param_y_predicted = lm.predict(param_X_test)
        test_f_hat[p,i] = test_param_y_predicted[0]  # prediction of ground truth y

# test_f_hat dimension = len(selected_params) x num_iterations
# np.mean(W,d) = average W in d dimension, where d = 0 = first dimension 
test_E_f_hat = np.mean(test_f_hat, 1)  # average over 2nd dimension or num_iterations
                                       # test_E_f_hat dimension = len(selected_params)
print(test_E_f_hat)
```

```python
# Plot distribution of f_hat for 3 parameters, along with E_f_hat and ground_truth_y
max_heights = [0.1, 0.1, 0.1]    # can choose other  values
colors = np.array(['tab:green', 'tab:purple', 'tab:orange'])

fig, axs = plt.subplots(len(selected_params), figsize=(12, 6), sharex=True)
for k in range(len(selected_params)):    
    # plot distribution of f_hat
    axs[k].hist(test_f_hat[k], density=True, color=colors[k], alpha=0.6, label=r'$\hat{f}$')   
    
    #plot E_f_hat
    axs[k].plot([test_E_f_hat[k], test_E_f_hat[k]], [0, max_heights[k]], c='k', linewidth=2.0, label=r'$\mathbb{E}[\hat{f}]$')

    # plot ground_truth_y
    axs[k].plot([ground_truth_y, ground_truth_y], [0, max_heights[k]], 'r', linewidth=2.0, label=r'$ground\_truth\_y$')

    # set title and legend
    axs[k].title.set_text('Train percent = {0:.0f}%, #Parameters = {1}'.format(train_percent*100, selected_params[k]))
    axs[k].legend(fontsize=12)
    
plt.show()
```

### Bias-Variance Tradeoff

```python
print(len(input_vars))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
test_percent = 0.2
main_X_train, X_test, main_y_train, y_test = train_test_split(X, y, test_size=test_percent) 

from sklearn.preprocessing import StandardScaler
scaled_main_X_train = StandardScaler().fit_transform(main_X_train)
scaled_X_test = StandardScaler().fit_transform(X_test)

interval = 1       # can choose other values
start_param = 2    # can choose other values
max_param = len(input_vars)   # can choose other  values
param_list = np.arange(start_param, max_param+1, interval).tolist()
print(param_list)

num_iterations = 1000  # can choose other  values

train_size = int(len(main_X_train)*train_percent)
test_size = len(X_test)
print(train_size)
print(test_size)

train_squared_error = np.zeros((len(param_list), num_iterations))  # shape: Params x Iterations
test_f_hat = np.zeros((len(param_list), num_iterations, test_size))  # shape: Params x Iterations x Test_size

for i in range(num_iterations):
    # Shuffle and select training set with specific train_size
    rand_index = np.random.permutation(len(main_X_train))[:train_size]  
    idx = main_y_train.index[rand_index]
    scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train.loc[idx]

    for p in range(len(param_list)):
        # Get the leftmost num_params columns to be inputs of linear model
        num_params = param_list[p]
        param_X_train = scaled_X_train[:,0:num_params]
        param_X_test = scaled_X_test[:,0:num_params]

        # Derive the model with the specified inputs
        lm = LinearRegression()
        lm.fit(param_X_train, y_train)

        # Make predictions for training and test sets
        train_param_y_predicted = lm.predict(param_X_train)
        test_param_y_predicted = lm.predict(param_X_test)

        # Compute MSE for all samples in training set
        train_squared_error[p, i] = np.mean((y_train.tolist() - train_param_y_predicted) ** 2) # shape: Params x Iterations

        # Collect predictions of test sets to compute MSE for test sets later
        test_f_hat[p, i, :] = test_param_y_predicted  # shape: Params x Iterations x Test_size

print('Done')
```

```python
# train_squared_error dimension = # shape: Params x Iterations
# Get average MSE for all iterations
avg_train_squared_error = np.mean(train_squared_error, 1)  # shape: Params

# test_f_hat dimension = Params x Iterations x Test_size
# Get average f_hat = E[f_hat] by avearaing over iterations
test_E_f_hat = np.mean(test_f_hat, 1)   # shape: Params x Test_size

# test_f_hat dimension = Params x Iterations x Test_size
# Compute test_squared_error = ||y - f_hat ||^2  for each iteration, for each sample in the test set
test_all_squared_error = (y_test.tolist() - test_f_hat) ** 2  # shape: Params x Iterations x Test_size
# Averaging test_squared_error for all iterations, for each sample in the test set  
test_sample_squared_error = np.mean(test_all_squared_error, 1)   # shape: Params x Test_size
# Compute average test_squared_error for all samples in the test set
avg_test_squared_error = np.mean(test_sample_squared_error, 1)    # shape: Params

# test_E_f_hat dimension = Params x Test_size
# Compute bias_squared = || y - E[f_hat] ||^2 for each sample in the test set
test_sample_bias_squared = (y_test.tolist() - test_E_f_hat) ** 2   # shape: Params x Test_size
# Compute average bias_squared for all samples in the test set
avg_test_bias_squared = np.mean(test_sample_bias_squared, 1)  # shape: Params

# test_f_hat dimension = Params x Iterations x Test_size
# Compute Var[f_hat] for each sample in the test set
test_sample_var_f_hat = np.var(test_f_hat, 1)   # shape: Params x Test_size
# Compute average bias_squared for all samples in the test set
avg_test_var_f_hat = np.mean(test_sample_var_f_hat, 1)  # shape: Params

print(avg_train_squared_error)
print(avg_test_squared_error)
print(avg_test_bias_squared)
print(avg_test_var_f_hat)
```

```python
# Plot bias-variance 
fig, ax = plt.subplots(3,1)
ax[0].plot(param_list, avg_test_bias_squared, 'y-.', label=r'bias squared: $(\mathbb{E}[\hat{f}] - y)^2$')  
ax[0].set_xticks(param_list)
ax[0].set_xticklabels(param_list, rotation=60)
ax[0].set_xlabel('Number of Parameters', size=10)
ax[0].set_ylabel('Squared Error', size=10)
ax[0].legend(loc='upper right', fontsize=10)
ax[0].set_title('Train percent = '+'{0:.0f}'.format(train_percent*100)+'%', fontsize=10)

ax[1].plot(param_list, avg_test_var_f_hat, 'b:', label=r'$var(\hat{f})$')  
ax[1].set_xticks(param_list)
ax[1].set_xticklabels(param_list, rotation=60)
ax[1].set_xlabel('Number of Parameters', size=10)
ax[1].set_ylabel('Squared Error', size=10)
ax[1].legend(loc='upper right', fontsize=10)

ax[2].plot(param_list, avg_test_bias_squared, 'y-.', label=r'bias squared: $(\mathbb{E}[\hat{f}] - y)^2$')  
ax[2].plot(param_list, avg_test_var_f_hat, 'b:', label=r'$var(\hat{f}$')  
ax[2].set_xticks(param_list)
ax[2].set_xticklabels(param_list, rotation=60)
ax[2].set_xlabel('Number of Parameters', size=10)
ax[2].set_ylabel('Squared Error', size=10)
ax[2].legend(loc='upper right', fontsize=10)

plt.show()
```

```python
# Plot training and testing errors
fig, ax = plt.subplots(3,1)
ax[0].plot(param_list, avg_test_squared_error, 'r', label='test error') 
ax[0].set_xticks(param_list)
ax[0].set_xticklabels(param_list, rotation=60)
ax[0].set_xlabel('Number of Parameters', size=10)
ax[0].set_ylabel('Squared Error', size=10)
ax[0].legend(loc='upper right', fontsize=10)
ax[0].set_title('Train percent = '+'{0:.0f}'.format(train_percent*100)+'%', fontsize=10)

ax[1].plot(param_list, avg_train_squared_error, 'g--', label='training error') 
ax[1].set_xticks(param_list)
ax[1].set_xticklabels(param_list, rotation=60)
ax[1].set_xlabel('Number of Parameters', size=10)
ax[1].set_ylabel('Squared Error', size=10)
ax[1].legend(loc='upper right', fontsize=10)

ax[2].plot(param_list, avg_test_squared_error, 'r', label='test error') 
ax[2].plot(param_list, avg_train_squared_error, 'g--', label='training error') 
ax[2].set_xticks(param_list)
ax[2].set_xticklabels(param_list, rotation=60)
ax[2].set_xlabel('Number of Parameters', size=10)
ax[2].set_ylabel('Squared Error', size=10)
ax[2].legend(loc='upper right', fontsize=10)

plt.show()
```
