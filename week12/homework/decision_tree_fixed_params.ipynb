### For a specific groundtruth y, Create (1) Distribution of f_hat, and find (2) a Value of E(f_hat)

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
test_percent = 0.2
main_X_train, X_test, main_y_train, y_test = train_test_split(X, y, test_size=test_percent)

from sklearn.preprocessing import StandardScaler
scaled_main_X_train = StandardScaler().fit_transform(main_X_train)
scaled_X_test = StandardScaler().fit_transform(X_test)

num_iterations = 1000

selected_percent = [0.2, 0.3, 0.4]
selected_train_size = [int(len(main_X_train)*percent) for percent in selected_percent]
print(selected_train_size)
test_size = len(X_test)

test_f_hat = np.zeros((len(selected_train_size), num_iterations))

# select the first value in y_test to be ground truth (can select any value in y_test)
ground_truth_y = y_test.tolist()[0]
print(ground_truth_y)

for t in range(len(selected_train_size)):
    train_size = selected_train_size[t]
    classifier = DecisionTreeClassifier()
    for i in range(num_iterations):
        rand_index = np.random.permutation(len(main_X_train))[:train_size]
        scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train[rand_index.tolist()]

        classifier.fit(scaled_X_train, y_train)
        test_param_y_predicted = classifier.predict(scaled_X_test)
        test_f_hat[t,i] = test_param_y_predicted[0]  # prediction of ground truth y

print('Done')
```

```python
print(test_f_hat)
```

```python
# Plot distribution of f_hat for 3 parameters, along with Mode_f_hat and ground_truth_y
max_heights = [1, 1, 1]   # can choose other  values

num_classes = 5     # adjust this value for your dataset
bins = np.arange(0, num_classes+1,1)

test_E_f_hat = []
for k in range(len(selected_train_size)):
    temp_mode = np.bincount(test_f_hat[k].astype(int)).argmax()
    test_E_f_hat.append(temp_mode)

colors = np.array(['tab:green', 'tab:purple', 'tab:orange'])
fig, axs = plt.subplots(len(selected_train_size), figsize=(12, 6))
for k in range(len(selected_train_size)):
    # plot distribution of f_hat
    axs[k].hist(test_f_hat[k], bins=bins, color=colors[k], density=True, alpha=0.6, label=r'$\hat{f}$')

    #plot E_f_hat
    axs[k].plot([test_E_f_hat[k], test_E_f_hat[k]], [0, max_heights[k]], c='k', linewidth=2.0, label=r'$Mode[\hat{f}]$')

    # plot ground_truth_y
    axs[k].plot([ground_truth_y, ground_truth_y], [0, max_heights[k]], 'r', linewidth=2.0, label=r'$ground\_truth\_y$')

    # set title and legend
    axs[k].title.set_text('#Parameters = {0}, Train Percent = {1:.0f}%'.format(num_params, selected_percent[k]*100))
    axs[k].legend(fontsize=12)

plt.show()
```

### Bias-Variance Tradeoff

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
test_percent = 0.2
main_X_train, X_test, main_y_train, y_test = train_test_split(X, y, test_size=test_percent)

from sklearn.preprocessing import StandardScaler
scaled_main_X_train = StandardScaler().fit_transform(main_X_train)
scaled_X_test = StandardScaler().fit_transform(X_test)

num_iterations = 1000      # can choose other  values

interval = 0.05             # can choose other  values
start_percent = 0.2         # can choose other  values
end_percent = 0.95             # can choose other  values
train_percent_list = np.arange(start_percent, end_percent, interval)
train_size_list = [int(len(main_X_train)*percent) for percent in train_percent_list]
print(train_size_list)

test_size = len(X_test)
print(test_size)

train_sample_error = np.zeros((len(train_size_list), num_iterations))  # shape: Percents x Iterations
test_f_hat = np.zeros((len(train_size_list), num_iterations, test_size))  # shape: Percents x Iterations x Test_size

for s in range(len(train_size_list)):
    train_size = train_size_list[s]
    classifier = DecisionTreeClassifier()

    for i in range(num_iterations):
        # Shuffle and select training set with specific train_size
        rand_index = np.random.permutation(len(main_X_train))[:train_size]
        #idx = main_y_train.index[rand_index]
        #scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train.loc[idx]
        scaled_X_train, y_train = scaled_main_X_train[rand_index.tolist()], main_y_train[rand_index.tolist()]

        classifier.fit(scaled_X_train, y_train)

        # Make predictions for training and test sets
        train_param_y_predicted = classifier.predict(scaled_X_train)
        test_param_y_predicted = classifier.predict(scaled_X_test)

        # Compute Average Error for all samples in training set
        # Error in classification = Incorrect classification
        # Note: train_sample_error dimesion = shape: Percents x Iterations
        train_sample_error[s, i] = sum([1 for i in range(train_size) if y_train.tolist()[i] != train_param_y_predicted[i]]) / train_size

        # Collect predictions of test sets to compute error for test sets later
        test_f_hat[s, i, :] = test_param_y_predicted  # shape: Percents x Iterations x Test_size

print('Done')
```

```python
#variance of the 0-1 loss is defined as the probability that the predicted label does not match the main prediction E[F_hat]
# E_f_hat = mode of f_hat

# train_sample_error dimension = # shape: Params x Iterations
# Get average MSE for all iterations
avg_train_error = np.mean(train_sample_error, 1)  # shape: Params
avg_train_error

test_sample_error = np.zeros((len(train_size_list), test_size))
test_E_f_hat = []
test_sample_bias = np.zeros(len(train_size_list))
test_sample_var_f_hat = np.zeros((len(train_size_list), test_size))
for s in range(len(train_size_list)):
    temp_E_f_hat = []
    test_sample_bias_count = 0

    for t in range(test_size):
        test_predictions = test_f_hat[s,:,t]  # shape: num_iterations
        test_predictions = test_predictions.astype(int)

        test_sample_error[s,t] = sum([1 for prediction in test_predictions if prediction != y_test[t]]) / num_iterations

        # find mode among all iterations, and assign it to be E[f_hat]
        E_f_hat = np.bincount(test_predictions).argmax()
        temp_E_f_hat.append(E_f_hat)

        test_sample_var_f_hat[s,t] = sum([1 for prediction in test_predictions if prediction != E_f_hat]) / num_iterations

    test_sample_bias[s] = sum([1 for i in range(len(y_test)) if temp_E_f_hat[i] != y_test[i]])  / test_size

# test_sample_error dimension = len(train_size_list) x test_size
avg_test_error = np.mean(test_sample_error, 1)  # shape: Percents

# test_sample_var_f_hat dimension = len(train_size_list) x test_size
avg_test_var_f_hat = np.mean(test_sample_var_f_hat, 1)  # shape: Percents

print(avg_train_error)
print(avg_test_error)
print(test_sample_bias)
print(avg_test_var_f_hat)
```

```python
# Plot bias-variance
x_list = [percent*100 for percent in train_percent_list]
xtick_labels = ['{0:.0f}'.format(percent*100) for percent in train_percent_list]

fig, ax = plt.subplots(3,1)
ax[0].plot(x_list, test_sample_bias, 'y-.', label=r'bias: $||\hat{f}] \neq y||$')
ax[0].set_xticks(x_list)
ax[0].set_xticklabels(xtick_labels, rotation=60)
ax[0].set_xlabel('Percent of Training Set', size=10)
ax[0].set_ylabel('Squared Error', size=10)
ax[0].legend(fontsize=10)
ax[0].set_title('Number of Parameters = '+str(num_params), fontsize=10)

ax[1].plot(x_list, avg_test_var_f_hat, 'b:', label=r'$var(\hat{f})$')
ax[1].set_xticks(x_list)
ax[1].set_xticklabels(xtick_labels, rotation=60)
ax[1].set_xlabel('Percent of Training Set', size=10)
ax[1].set_ylabel('Squared Error', size=10)
ax[1].legend(fontsize=10)

ax[2].plot(x_list, test_sample_bias, 'y-.', label=r'bias: $||\hat{f}] \neq y||$')
ax[2].plot(x_list, avg_test_var_f_hat, 'b:', label=r'$var(\hat{f})$')
ax[2].set_xticks(x_list)
ax[2].set_xticklabels(xtick_labels, rotation=60)
ax[2].set_xlabel('Percent of Training Set', size=10)
ax[2].set_ylabel('Squared Error', size=10)
ax[2].legend(fontsize=10)

plt.show()
```
