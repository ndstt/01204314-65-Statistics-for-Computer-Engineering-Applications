```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

Scipy.stats provides functions to work with general distributions, such as uniform, normal, t, chi-squared,etc.

See below for examples.

## Uniform Distribution

```python
# Standard uniform between x \in [0, 1]
x = np.linspace(st.uniform.ppf(0.01),
                st.uniform.ppf(0.99), 100)
plt.plot(x, st.uniform.pdf(x),
       'r-', lw=5, alpha=0.6, label='uniform pdf')
```

```python
# Scaling uniform distribution
x = np.linspace(st.uniform.ppf(0.01, scale=2),
                st.uniform.ppf(0.99, scale=2), 100)
plt.plot(x, st.uniform.pdf(x, scale=2),
       'r-', lw=5, alpha=0.6, label='uniform pdf')
```

```python
# Scaling uniform distribution + Setting minimum outcomes
x = np.linspace(st.uniform.ppf(0.01, loc=5, scale=2),
                st.uniform.ppf(0.99, loc=5, scale=2), 100)
plt.plot(x, st.uniform.pdf(x, loc=5, scale=2),
       'r-', lw=5, alpha=0.6, label='uniform pdf')
```

```python
# Derive probability of outcome x
# uniform.pdf(x, loc, scale)
prob = st.uniform.pdf(6.0, loc=5, scale=2)
prob
```

```python
# Derive cumulative probability of outcome x
# uniform.cdf(x, loc, scale)
cumulative_prob = st.uniform.cdf(5.5, loc=5, scale=2)
cumulative_prob
```

```python
# Derive inverse CDF = Get outcome x when cumulative probability cp is entered
# uniform.ppf(cp, loc, scale)
x = st.uniform.ppf(0.25, loc=5, scale=2)
x
```

```python
# Derive random outcomes with size n according to specified uniform distribution
# uniform.rvs(loc, scale, size=n)
x = st.uniform.rvs(loc=5, scale=2, size=20)
x
```

```python
# Derive basic statistics of specified uniform distribution
# mean, var = st.uniform.stats(moments='mv')
mean, var, skew, kurt = st.uniform.stats(moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

## Normal Distribution

```python
# Standard normal distribution where mean = 0, sd =1
x = np.linspace(st.norm.ppf(0.01),
                st.norm.ppf(0.99), 100)
plt.plot(x, st.norm.pdf(x),
       'r-', lw=5, alpha=0.6, label='normal pdf')
```

```python
# Scaling normal distribution
x = np.linspace(st.norm.ppf(0.01, scale=3),
                st.norm.ppf(0.99, scale=3), 100)
plt.plot(x, st.norm.pdf(x, scale=3),
       'r-', lw=5, alpha=0.6, label='normal pdf')
```

```python
# Scaling normal distribution + Setting minimum outcomes
x = np.linspace(st.norm.ppf(0.01, loc=5, scale=3),
                st.norm.ppf(0.99, loc=5, scale=3), 100)
plt.plot(x, st.norm.pdf(x, loc=5, scale=3),
       'r-', lw=5, alpha=0.6, label='normal pdf')
```

```python
# Derive probability of outcome x
# norm.pdf(x, loc, scale)
prob = st.norm.pdf(0.0)
prob
```

```python
# Derive cumulative probability of outcome x
# norm.cdf(x, loc, scale)
cumulative_prob = st.norm.cdf(0.0)
cumulative_prob
```

```python
cumulative_prob = st.norm.cdf(1.96)
cumulative_prob
```

```python
cumulative_prob = st.norm.cdf(1.6449)
cumulative_prob
```

```python
# Derive inverse CDF = Get outcome x when cumulative probability cp is entered
# norm.ppf(cp, loc, scale)
x = st.norm.ppf(0.5)
x
```

```python
x = st.norm.ppf(0.95)
x
```

```python
x = st.norm.ppf(0.975)
x
```

```python
# Derive random outcomes with size n according to specified normal distribution
# norm.rvs(loc, scale, size=n)
x = st.norm.rvs(size=20)
x
```

```python
# Derive basic statistics of specified normal distribution
# mean, var = st.norm.stats(moments='mv')
mean, var, skew, kurt = st.norm.stats(moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```
