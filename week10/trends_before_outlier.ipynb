```python
df_regions = pd.read_excel(os.path.join(data_path, "countries_regions.xlsx"))
df_regions.reset_index(inplace=True)
df_regions.rename(columns={"CountryName": "Country"}, inplace=True)

# โหลดข้อมูลเศรษฐกิจ (Economics)
df_inflation = pd.read_csv(os.path.join(econ_path, "FP.CPI.TOTL.ZG_ts.csv"))
df_debt = pd.read_csv(os.path.join(econ_path, "GC.DOD.TOTL.GD.ZS_ts.csv"))
df_tax = pd.read_csv(os.path.join(econ_path, "GC.TAX.TOTL.GD.ZS_ts.csv"))
df_expense = pd.read_csv(os.path.join(econ_path, "GC.XPN.TOTL.GD.ZS_ts.csv"))
df_gdp = pd.read_csv(os.path.join(econ_path, "NY.GDP.MKTP.CD_ts.csv"))

# โหลดข้อมูลโครงสร้างพื้นฐาน (Infrastructure)
df_electricity = pd.read_csv(os.path.join(infra_path, "EG.ELC.ACCS.ZS_ts.csv"))
df_internet = pd.read_csv(os.path.join(infra_path, "IT.NET.BBND.P2_ts.csv"))
df_hospital = pd.read_csv(os.path.join(infra_path, "SH.MED.BEDS.ZS_ts.csv"))

# โหลดข้อมูลประชากร (Population)
df_life_expectancy = pd.read_csv(os.path.join(pop_path, "SP.DYN.LE00.IN_ts.csv"))
df_pop_65_up = pd.read_csv(os.path.join(pop_path, "SP.POP.65UP.TO.ZS_ts.csv"))
df_pop_working_age = pd.read_csv(os.path.join(pop_path, "SP.POP.1564.TO.ZS_ts.csv"))
df_total_population = pd.read_csv(os.path.join(pop_path, "SP.POP.TOTL_ts.csv"))
```

```python
def wide_to_long_format(df, valuename):
    return df.melt(id_vars=["Year"], var_name="Country", value_name=valuename)

def merge_with_regions(df):
    return df.merge(df_regions, on="Country", how="left")

def rolling_avg(df,valuename):
    df = wide_to_long_format(df, valuename)
    df = merge_with_regions(df)
    df = df.groupby(["Region", "Year"])[valuename].mean().reset_index()
    df["Rolling_Avg"] = df.groupby("Region")[valuename].transform(lambda x: x.rolling(5, min_periods=1).mean())
    return df
```

```python
def show_trend(df,Title):
    for region in df["Region"].dropna().unique():
        region_data = df[df["Region"] == region]
        plt.plot(region_data["Year"], region_data["Rolling_Avg"], label=region)
    plt.title("5-Year Rolling Average of Region's " + Title)
    plt.xlabel("Year")
    plt.ylabel(Title)
    plt.legend(fontsize=7, loc='best', ncol=2)
```
